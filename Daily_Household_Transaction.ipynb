```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df["Date"] = pd.to_datetime(
    df["Date"],
    dayfirst=True,
    errors="coerce"
)
```

```python
df = df.dropna(subset=["Date"])
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df,x="Mode",order=df['Mode'].value_counts().iloc[:3].index)
plt.title("Top payment models")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df,x="Category",order=df['Category'].value_counts().iloc[:5].index)
plt.title("Top 5 expense categories")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x=df['Income/Expense'])
plt.title("Income vs Expense")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(
    data=df,
    x="Amount",
    y="Category",
    order=df["Category"].value_counts().iloc[:5].index
)
plt.title("Amount Distribution by Category")
plt.show()
```

```python
sns.boxplot(data=df, x="Income/Expense", y="Amount")
plt.title("Income vs Expense Amount Distribution")
plt.show()
```

```python
daily_expense = df.groupby(df["Date"].dt.date)["Amount"].sum()

plt.figure(figsize=(12,6))
plt.plot(daily_expense)
plt.title("Daily Transaction Amount Trend")
plt.xlabel("Date")
plt.ylabel("Amount")
plt.xticks(rotation=45)
plt.show()
```
